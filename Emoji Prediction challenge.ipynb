```python
### Loading the training and testing set

import pandas as pd
```

```python
xtr = pd.read_csv('datasets/emojify_train_x.csv', header=None).values
xte = pd.read_csv('datasets/emojiy_test_x.csv', header=None).values
ytr = pd.read_csv('datasets/Emojify_Y_train.csv', header=None).values
yte = pd.read_csv('datasets/emojiy_y_test.csv', header=None).values
```

```python
print(ytr)
```

```python
## Preproces the data


def get_array(x):
    x_tr = []
    for i in range(x.shape[0]):
        x_ = x[i, 0].split("' '")
        for sent in x_:
            s = sent.replace("'", "")
            x_tr.append(s)
    
    return np.array(x_tr)
            

def get_y(y):

    y_tr = []

    for i in range(y.shape[0]):
        y_ = y[i, 0].split(" ")
        for w in y_:
            if w != '':
                y_tr.append(int(w))
            
    return np.array(y_tr)
```

```python
import numpy as np

x_train = get_array(xtr)
x_test = get_array(xte)
```

```python
y_train = get_y(ytr)
y_test = get_y(yte)
```

```python
print(x_train, y_train)
print(x_train.shape, y_train.shape)
```

```python
### Glove dictionary

embedding_dict = {}

with open('datasets/glove.6B.50d.txt', encoding='utf-8') as f:
    
    for line in f:
        values = line.split()
        word = values[0]
        array = np.asarray(values[1:], dtype='float')
        embedding_dict[word] = array
```

```python
embedding_dict["sleep"]
```

```python
def embedding_output(X):
    maxLen = 10
    embed_dim = 50
    
    out = np.zeros((X.shape[0], maxLen, embed_dim))
    
    for ix in range(X.shape[0]):
        words = X[ix].split()
        for ij in range(len(words)):
            try:
                ar = embedding_dict[words[ij].lower()]
                out[ix, ij] = ar
            except:
                out[ix, ij] = np.zeros((50, ))
            
    return out
```

```python
embedding_x_train = embedding_output(x_train)
embedding_x_test = embedding_output(x_test)
```

```python
print(embedding_x_train.shape, embedding_x_test.shape)
```

```python
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import *
```

```python
y_train = np_utils.to_categorical(y_train, num_classes=5)
y_test = np_utils.to_categorical(y_test, num_classes=5)
```

```python
### Making the model

model = Sequential()
model.add(LSTM(64, input_shape=(10, 50)))
model.add(Dropout(0.5))
model.add(Dense(5, activation='softmax'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
from keras.callbacks import ModelCheckpoint
from keras.models import load_model


checkpoint = ModelCheckpoint("best_model.h5", monitor='val_loss', verbose=True, save_best_only=True)
```

```python


hist = model.fit(embedding_x_train, y_train, shuffle=True, validation_split=.2, epochs=100, batch_size=32, callbacks=[checkpoint])
```

```python
best_model = load_model("best_model.h5")
```

```python
Acc = model.evaluate(embedding_x_test, y_test)[1]
```
